```python
# online judge - 10327
# WA

#include <iostream>
#include <cmath>
#include <algorithm>
#include <cstdio>
using namespace std;

int bubble_sort(long long num[], int n)
{
    int count=0;
    for(int i = 0; i < n; ++i)
    {
        for(int j = 0; j < n-i-1; ++j)
        {
            if (num[j] > num[j+1])
            {
                //printf("%lld %lld\n", num[j-1], num[j]);
                long long s = num[j-1];
                num[j-1] = num[j];
                num[j] = s;
                ++count;
                //printf("%d\n", count);
            }
        }
    }
    return count;
}

int main()
{
    long long a[1000];
    int i, n;
    
    while(scanf("%d\n", &n)!=EOF)
    {
        for(i=0; i < n; ++i)
        {
            scanf("%lld ", &a[i]);
        }
        
    
    
        int result = bubble_sort(a, n);
    
        printf("Minimum exchange operations : %d\n", result);
    
    }
    
    return 0;
}
```

```python
# online judge - 10347
# RTE
def flip_sort(numbers):
    flip_count = 0
    for i in range(len(numbers)):
        for j in range(1, len(numbers)-i):
            if numbers[j-1] > numbers[j]:
                numbers[j], numbers[j-1] = numbers[j-1], numbers[j]
                flip_count += 1
                
    return flip_count
                
if __name__ == "__main__":
    while True:
        try:
            n = int(input())
            numbers = [int(x) for x in input().split()]
            
            print("Minimum exchange operations : {}".format(flip_sort(numbers)))    
            
        except EOFError:
            break
```

```python
# online judge - 10879

#include <iostream>
#include <cmath>

using namespace std;

int main()
{
    int N, K, N1, N2, i, j;
    
    cin>>N;
    
    for(i=0; i<N; ++i)
    {
        cin >> K;
        
        N1 = N2 = 0;
        
        for(j=2; j<=sqrt(K); ++j)
        {
            if (j*(K/j)==K) {
                N1 = j;
                break;
        
            }
        }
        
        for(j=N1+1; j<=sqrt(K); ++j)
        {
            if ((j*(K/j)==K) && (j!=(K/N1))) {
                N2 = j;
                break;
        
            }
        }
        
        cout << "Case #" << i+1 <<": " << K << " = " << N1 << " * " << K/N1 
        << " = " << N2 << " * " << K/N2 << endl;
    }
    
    
    return 0;
}
```

```python
# online judge - 10879
# TLE
#include <stdio.h>

int main()
{
    long int K, q, j, p, factors[10000000];
    int N, i, count;
    
    scanf("%d\n", &N);
    
    for(i=0; i<N; ++i)
    {
        scanf("%ld\n", &K);
        
        printf("Case #%d: %ld", i+1, K);
        
        j = 0;
        
        count = 0;
        for(p=2; p<=K/2; ++p)
        {
            if (K % p == 0)
            {
                factors[j] = p;
                
                for (q=0; q < j-1; ++q) 
                {
                    if (factors[q] * factors[j] == K)
                    {
                        printf(" = %ld * %ld", factors[q], factors[j]);
                        ++count;
                        if (count == 2)
                            goto end; 
                    }
                }
                ++j; 
            }    
        }
        
        end:
        printf("\n");
        
    }
    
    return 0;
}
```
